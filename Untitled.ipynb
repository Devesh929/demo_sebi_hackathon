```python
def subsets(numbers):
	if numbers == []:
		return [[]]
	x = subsets(numbers[1:])
	return x + [[numbers[0]] + y for y in x]

# wrapper function
def subsets_of_given_size(numbers, n):
	return [x for x in subsets(numbers) if len(x)==n]

if __name__ == '__main__':
    numbers =[1,2,5,10,20,50,100]
    n = 4
    l4=subsets_of_given_size(numbers, 4)
    l5=subsets_of_given_size(numbers, 5)
    l6=subsets_of_given_size(numbers, 6)
    l7=subsets_of_given_size(numbers, 7)
    

```
